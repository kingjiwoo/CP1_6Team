```python
import matplotlib.pyplot as plt
import numpy as np

# 그림 사이즈, 바 굵기 조정
fig, ax = plt.subplots(figsize=(12,6))
bar_width = 0.25

# 연도가 4개이므로 0, 1, 2, 3 위치를 기준으로 삼음
index = np.arange(4)

# 각 연도별로 3개 샵의 bar를 순서대로 나타내는 과정, 각 그래프는 0.25의 간격을 두고 그려짐
b1 = plt.bar(index, df_vs['instrumentalness'], bar_width, alpha=0.4, color='red', label='shop A')

b2 = plt.bar(index + bar_width, df_vs['popularity'], bar_width, alpha=0.4, color='blue', label='shop B')

b3 = plt.bar(index + 2 * bar_width, df_vs['danceability'], bar_width, alpha=0.4, color='green', label='shop C')

# x축 위치를 정 가운데로 조정하고 x축의 텍스트를 year 정보와 매칭
plt.xticks(np.arange(bar_width, 4 + bar_width, 1), index)

# x축, y축 이름 및 범례 설정
plt.xlabel('feature', size = 13)
plt.ylabel('number', size = 13)
plt.legend()
plt.show()
```
